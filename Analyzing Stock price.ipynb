```python
def top_com_week(week, value, top_num):
    Week_specific = group_week_com[group_week_com.WeekID == week]  
    sorted_df = Week_specific.sort_values(value, ascending= False).head(10)
    plt.figure(figsize = (16,7))
    plt.bar( sorted_df['Scrip'], sorted_df[value], color='blue', linestyle='solid', linewidth = 3, width = .5),
    plt.title('Top Compianies on specific week', fontsize = 15),
    plt.xlabel('Companies',fontsize = 15),
    plt.ylabel(value, fontsize = 15),
    for index, val in enumerate(sorted_df[value]):
        plt.text(index, val+(val//30), str(val), ha = 'center')
        
    plt.show()
```
